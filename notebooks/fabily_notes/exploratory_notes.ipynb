Partindo do ponto em que os dados já estão limpos, chegamos à etapa de "exploração dos dados". Primeiramente, pretendo entender quais são as colunas, que dados elas trazem e suas necessidades para a análise como um todo. Para isso podemos usar alguns comandos:

```python
import numpy as np
import pandas as pd

df = pd.read_csv("../../data/processed/olist_super_dataset.csv")
```

```python
df.info()
```

Como podemos observar, temos muitas ocorrências/registros para analisar. Isso traz os dois lados da moeda, tanto a abrangência dos dados em larga escala quanto o poder computacional necessário para trabalhar com todos esses dados...

Outra coisa à se observar é o fato de termos algumas linhas nulas, conforme vemos em:

```python
tot_lines = len(df.index)
null_lines = df.isnull().any(axis=1).sum()

print(f"""Quantidade de linhas nulas: {null_lines}
Quantidade de células nulas: {df.isnull().sum().sum()}
Porcentagem aem relação ao total: {(null_lines*100)/tot_lines:.2f}%""")
```

Com essa informação, sabemos que algumas linhas possuem mais de uma célula nula. Agora temos de decidir o que fazer com essas linhas... Eu irei optar por retirá-las - uma vez que elas representam apenas 2.15% do total. Para retirá-las podemos usar a abordagem:

```python
df.dropna(axis=0, how='any', inplace=True)

tot_lines = len(df.index)
null_lines = df.isnull().any(axis=1).sum()

print(f"""Quantidade de linhas nulas: {null_lines}
Quantidade de células nulas: {df.isnull().sum().sum()}
Porcentagem aem relação ao total: {(null_lines*100)/tot_lines:.2f}%""")
```
